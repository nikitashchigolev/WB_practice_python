```python
# Процедура для преобразования в двоичную кучу поддерева с корневым узлом i, что является индексом в arr[]. n - размер кучи
def heapify(arr, n, i):
    largest = i # Initialize largest as root
    l = 2 * i + 1   # left = 2*i + 1
    r = 2 * i + 2   # right = 2*i + 2

  # Проверяем существует ли левый дочерний элемент больший, чем корень

    if l < n and arr[i] < arr[l]:
        largest = l

    # Проверяем существует ли правый дочерний элемент больший, чем корень

    if r < n and arr[largest] < arr[r]:
        largest = r

    # Заменяем корень, если нужно
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i] # свап

        # Применяем heapify к корню.
        heapify(arr, n, largest)

# Основная функция для сортировки массива заданного размера
def heapSort(arr):
    n = len(arr)

    # Построение max-heap.
    for i in range(n//2, -1, -1):
        heapify(arr, n, i)

    # Один за другим извлекаем элементы
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i] # свап 
        heapify(arr, i, 0)
```

```python
import csv

# Открываем CSV файл для чтения
with open('drinks.csv', newline='') as csvfile:
    # Создаем объект для чтения CSV
    csvreader = csv.reader(csvfile)

    header = next(csvreader)

    column1 = []
    column2 = []
    column3 = []
    column4 = []
    column5 = []

    for row in csvreader:
        column1.append(row[0])
        column2.append(row[1])
        column3.append(row[2])
        column4.append(row[3])
        column5.append(row[4])

print(column1)
print(column2)
print(column3)
print(column4)
print(column5)
print(header)
print(len(column1),
     len(column2),
     len(column3),
     len(column4),
     len(column5))
```

```python
data = np.zeros(
    1894,
    dtype={
        "names": ['movie', 
                  'type', 
                  'word', 
                  'minutes_in'],
        "formats": ["U16", "U16", "U16", "f8"]
    }
)
print(data.dtype)
```

```python
data['movie'] = column1
data['type'] = column2
data['word'] = column3
data['minutes_in'] = column4

for _ in data:
    print(_)
```

```python
# Создание маски для прокрятий и летальных исходов
is_word = data['type'] == 'word'
is_death = data['type'] == 'death'

# Np.sum для подсчета прокрятий и летальных исходов
word_counts = np.sum(is_word)
death_counts = np.sum(is_death)

# общее количество записей для каждого фильма
movie_names = np.unique(data['movie'])
movie_word_counts = np.array([np.sum(is_word[data['movie'] == movie]) for movie in movie_names])
movie_death_counts = np.array([np.sum(is_death[data['movie'] == movie]) for movie in movie_names])

# отношение прокрятий к летальным исходам для каждого фильма
word_to_death_ratio = movie_word_counts / movie_death_counts

# результат
for i, movie in enumerate(movie_names):
    print(f"Фильм: {movie}")
    print(f"Количество прокрятий: {movie_word_counts[i]}")
    print(f"Количество смертей: {movie_death_counts[i]}")
    print(f"Отношение прокрятий к летальным исходам: {word_to_death_ratio[i]}")
    print()
```

```python
# Извлеките только строки с типом 'word'
words = data[is_word]['word']

# Используйте numpy.unique для подсчета частоты каждого прокрятия
unique_words, word_counts = np.unique(words, return_counts=True)

# результат
for word, count in zip(unique_words, word_counts):
    print(f"Слово: {word}")
    print(f"Частота: {count}")
    print()
```

```python
import matplotlib.pyplot as plt
# извлекаем только времена для прокрятий
word_minutes = data[is_word]['minutes_in']

# строим гистограмму
plt.hist(word_minutes, bins=20, edgecolor='k', alpha=0.7)
plt.title("Распределение времени между прокрятиями 'word'")
plt.xlabel("Время (минуты)")
plt.ylabel("Частота")
plt.show()
```
